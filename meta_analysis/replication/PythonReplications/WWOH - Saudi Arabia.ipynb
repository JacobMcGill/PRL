```python
import pandas as pd
import numpy as np

# List of variables to summarize
sum_vars = ["age", "college_deg", "children", "employed_now", "employed_wife", 
            "num_know_per", "num_mfs_per"]

# Labels for variables
var_labels = {
    "age": "Age",
    "college_deg": "College Degree (%)",
    "children": "Number of Children",
    "employed_now": "Employed (%)",
    "employed_wife": "Wife Employed (%)",
    "num_know_per": "Other Participants Known (%)",
    "num_mfs_per": "Other Participants with Mutual Friends (%)",
}

# Initialize dictionaries to store summary statistics
summary_stats = {}
for var in sum_vars:
    summary_stats[var] = {"all": {}, "0": {}, "1": {}}

# Loop through variables
for var in sum_vars:
    for i in [0, 1]:
        # Calculate mean and standard deviation for treatment and control
        summary_stats[var][str(i)]["mean"] = data.loc[data["condition2"] == i, var].mean()
        summary_stats[var][str(i)]["sd"] = data.loc[data["condition2"] == i, var].std()

    # Calculate mean and standard deviation for the full sample
    summary_stats[var]["all"]["mean"] = data[var].mean()
    summary_stats[var]["all"]["sd"] = data[var].std()

# Print summary statistics
for var in sum_vars:
    print(f"{var_labels[var]} & {summary_stats[var]['all']['mean']:.2f} & {summary_stats[var]['0']['mean']:.2f} & {summary_stats[var]['1']['mean']:.2f} \\\\")
    print(f" & ({summary_stats[var]['all']['sd']:.2f}) & ({summary_stats[var]['0']['sd']:.2f}) & ({summary_stats[var]['1']['sd']:.2f}) \\\\")

# Number of observations
count_all = len(data)
count_0 = data[data["condition2"] == 0].shape[0]
count_1 = data[data["condition2"] == 1].shape[0]

print(f"\nN & {count_all} & {count_0} & {count_1} \\\\")
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import bootstrapped.bootstrap as bs
import bootstrapped.stats_functions as bs_stats

# Assuming 'data' is your DataFrame
# Replace 'data' with the actual variable name you have in Python

# Create the 'update' variable
data['update'] = data['outside_wedge'] * (data['condition2'] == 1)

# Replace 'update' values for condition2 == 0
data.loc[data['condition2'] == 0, 'update'] = 0

# Model specifications
specs = [
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence',
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence + C(session)',
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence + C(session) + employed_wife + employed_now + education + children + num_know_per + num_mfs_per + age'
]

# Create a list to store p-values
p_values = []

# Set the number of bootstrap samples
num_bootstrap_samples = 1000

# Loop through the specifications
for i, spec in enumerate(specs, start=1):
    # Fit the regression model
    model = sm.OLS.from_formula(spec, data=data)
    result = model.fit()

    # Get the bootstrap p-value for the 'update' variable
    values = np.array(result.resid)
    bootstrap_result = bs.bootstrap(values, stat_func=bs_stats.mean, num_iterations=num_bootstrap_samples)
    p_value = bootstrap_result.lower_bound + (1 - bootstrap_result.upper_bound)

    # Append p-value to the list
    p_values.append(p_value)

# Display the bootstrap p-values
for i, p_value in enumerate(p_values, start=1):
    print(f"Model {i} Wild Bootstrap p-value for 'update': {p_value:.4f}")
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from permute.core import one_sample

# Assuming 'data' is your DataFrame
# Replace 'data' with the actual variable name you have in Python

# Create the 'update' variable
data['update'] = data['outside_wedge'] * (data['condition2'] == 1)

# Replace 'update' values for condition2 == 0
data.loc[data['condition2'] == 0, 'update'] = 0

# Model specifications
specs = [
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence',
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence + C(session)',
    'signed_up_number ~ update + outside_self + outside_others + outside_confidence + C(session) + employed_wife + employed_now + education + children + num_know_per + num_mfs_per + age'
]

# Create a list to store p-values
p_values = []

# Set the number of permutations
num_permutations = 1000

# Loop through the specifications
for i, spec in enumerate(specs, start=1):
    # Fit the regression model
    model = sm.OLS.from_formula(spec, data=data)
    result = model.fit()

    # Permutation test
    permuted_stats = one_sample(data['signed_up_number'], np.random.permutation(data['update']), reps=num_permutations)

    # Calculate the p-value
    obs_statistic = result.params['update'] / result.bse['update']
    p_value = np.sum(np.abs(permuted_stats) >= np.abs(obs_statistic)) / num_permutations

    # Append p-value to the list
    p_values.append(p_value)

# Display the permutation test p-values
for i, p_value in enumerate(p_values, start=1):
    print(f"Model {i} Permutation Test p-value for 'update': {p_value:.4f}")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col

# Regression model specifications
model1 = 'signed_up_number ~ condition2'
model2 = 'signed_up_number ~ condition2 + C(session)'
model3 = 'signed_up_number ~ condition2 + C(session) + ' + ' + '.join(self_columns) + ' + ' + ' + '.join(others_columns)
model4 = 'signed_up_number ~ condition2 + C(session) + ' + ' + '.join(self_columns) + ' + ' + ' + '.join(others_columns) + ' + employed_wife + employed_now + C(education) + children + num_know_per + num_mfs_per + age'

# Fit regressions
reg1 = sm.OLS.from_formula(model1, data=data).fit(cov_type='HC3')  # Robust standard errors
reg2 = sm.OLS.from_formula(model2, data=data).fit(cov_type='HC3')
reg3 = sm.OLS.from_formula(model3, data=data).fit(cov_type='HC3')
reg4 = sm.OLS.from_formula(model4, data=data).fit(cov_type='HC3')

# Extract p-values
p_values = [
    reg1.pvalues['condition2'],
    reg2.pvalues['condition2'],
    reg3.pvalues['condition2'],
    reg4.pvalues['condition2']
]

# Create a DataFrame for p-values
p_values_df = pd.DataFrame({"p-value (robust SE)": p_values})

# Print the p-values DataFrame
print(p_values_df)
```

```python
# Perform wild bootstrap for each model
wboot1 = reg1.get_robustcov_results().pvalues[0]
wboot2 = reg2.get_robustcov_results().pvalues[0]
wboot3 = reg3.get_robustcov_results().pvalues[0]
wboot4 = reg4.get_robustcov_results().pvalues[0]

# Create a DataFrame for wild bootstrap p-values
wboot_p_values_df = pd.DataFrame({"p-value (wild bootstrap)": [wboot1, wboot2, wboot3, wboot4]})

# Print the wild bootstrap p-values DataFrame
print(wboot_p_values_df)
```

```python
import pandas as pd
from mlxtend.evaluate import permutation_test
import statsmodels.api as sm

# Create observed values and expected values for each model
observed_values = [reg.model.endog for reg in [reg1, reg2, reg3, reg4]]
expected_values = [reg.fittedvalues for reg in [reg1, reg2, reg3, reg4]]

# Perform permutation test for each model
perm_test_results = [permutation_test(obs, exp, method='approximate', num_rounds=1000) for obs, exp in zip(observed_values, expected_values)]

# Extract p-values from permutation test results
perm_p_values = [result for result in perm_test_results]

# Create a DataFrame for permutation test p-values
perm_p_values_df = pd.DataFrame({"Model": [f"Model {i+1}" for i in range(4)],
                                  "Permutation Test P-Value": perm_p_values})

# Print the permutation test p-values DataFrame
print(perm_p_values_df)
```

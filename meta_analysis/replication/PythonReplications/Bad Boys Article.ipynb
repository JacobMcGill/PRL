```python
data = pd.read_csv("/Users/adishsundar/Desktop/Salem Center Research/Supplementary/Data/badboys_dataset.csv")
df = data
data.head()
```

```python
# Table 1 Replication - Everything is replicated properly
```

```python
import pandas as pd
import statsmodels.api as sm

# Filter rows where main is equal to 1
df_filtered = df[df['main'] == 1]

# Reshape the data with both 'coin' and 'treatment'
df_long = pd.melt(df_filtered, id_vars=['id', 'treatment'], value_vars=['coin1', 'coin2', 'coin3', 'coin4', 'coin5', 'coin6', 'coin7', 'coin8', 'coin9', 'coin10'], var_name='toss', value_name='coin')
df_long['toss'] = df_long['toss'].str.extract('(\d+)').astype(int) - 1

# Perform regression
model = sm.OLS(df_long['coin'], sm.add_constant(df_long[['treatment', 'toss']])).fit()

# Display regression results
print("Regression analysis: criminal identity and cheating")
print("(1)")
print("Dependent variable: heads = 1")

# Regression 1
print(model.summary())
```

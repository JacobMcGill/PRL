```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_stata('rationing_worker.dta')
```

```python
df
```

#### Table 4 Alternate Measures of Wages and Wage Contract

```python
# 1st column - hourly total wage

# Drop rows with missing values in 'hourlywagetot_w'
df_cleaned = df.dropna(subset=['hourlywagetot_w'])

# Filter the DataFrame based on the conditions specified in Stata code
df_filtered = df_cleaned[(df_cleaned['surv_el'] == 1) & 
                 (df_cleaned['sample_spillover'] == 1) & 
                 (df_cleaned['e_manwage'] == 1)]

# Regression with fixed effects and clustered standard errors
model = smf.mixedlm("hourlywagetot_w ~ treat + treat_peak",
                    data=df_filtered,
                    groups=df_filtered['vill_id'],
                    re_formula="~ round_id")

result = model.fit()

# Printing estimation results
print(result.summary())
```

```python
# 2nd column - hours per day

# Drop rows with missing values in 'hourlywagetot_w'
df_cleaned = df.dropna(subset=['hours'])

# Filter the DataFrame based on the conditions specified in Stata code
df_filtered = df_cleaned[(df_cleaned['surv_el'] == 1) & 
                 (df_cleaned['sample_spillover'] == 1) & 
                 (df_cleaned['e_manwage'] == 1) &
                 (df_cleaned['gotdailywage'] == 1)]

# Regression with fixed effects and clustered standard errors
model = smf.mixedlm("hours ~ treat + treat_peak",
                    data=df_filtered,
                    groups=df_filtered['vill_id'],
                    re_formula="~ round_id")

result = model.fit()

# Printing estimation results
print(result.summary())
```

#### Table 5 Employment Spillovers

```python
# 1st column - hired wage empl.

# Filter the DataFrame based on the conditions specified
df_filtered = df[(df['surv_el'] == 1) & 
                         (df['sample_spillover'] == 1)]

# Fit the regression model
model = smf.mixedlm("e_manwage ~ treat + treat_peak",
                    data=df_filtered,
                    groups=df_filtered['vill_id'],
                    re_formula="~ round_id")

result = model.fit()

# Printing estimation results
print(result.summary())
```

```python
# 2nd column - hired wage empl.

# Filter the DataFrame based on the conditions specified
df_filtered = df[(df['surv_el'] == 1) & 
                         (df['sample_spillover'] == 1)]

# Fit the regression model
model = smf.mixedlm("e_manwage ~ treat + treat_peak + bl_ldwagetot + bl_dwagetot_miss + bl_e_manwage + bl_e_manwage_miss",
                    data=df_filtered,
                    groups=df_filtered['vill_id'],
                    re_formula="~ round_id")

result = model.fit()

# Printing estimation results
print(result.summary())
```

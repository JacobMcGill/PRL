```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Define region codes
eastern_codes = {'SD', 'MD2', 'MD3', 'MD4', 'MW2', 'MW4', 'MZ3', 'MZ4', 'ZM2', 'ZM3', 'ZM4', 'ZW', 'ZW3', 'ZW4', 'ZW5', 'TZ2', 'TZ3', 'TZ4', 'ET4', 'KE', 'KE2', 'KE3', 'KE4', 'BU', 'RW2', 'RW4', 'UG', 'UG3', 'UG4', 'KM3', ''}
western_codes = {'BF2', 'BF3', 'BF4', 'ML', 'ML3', 'ML4', 'NI2', 'NI3', 'SN', 'SN2', 'SN4', 'GH', 'GH2', 'GH3', 'GH4', 'GN', 'LB', 'NG2', 'NG3', 'NG4', 'TG', 'TG3', 'BJ3', 'BJ4', 'CI3', 'CI4', 'CI5', 'GN3', 'GN4', ''}
central_codes = {'TD3', 'TD4', 'CM2', 'CM3', 'CM4', 'CF3', 'GA3'}
southern_codes = {'NM2', 'NM4', 'LS4', 'ZA3'}

# Create region assignments
df['REGION'] = 0
df.loc[df['v000'].isin(eastern_codes), 'REGION'] = 1
df.loc[df['v000'].isin(central_codes), 'REGION'] = 2
df.loc[df['v000'].isin(southern_codes), 'REGION'] = 3
df.loc[df['v000'].isin(western_codes), 'REGION'] = 4

# Basic regression analysis for Eastern Africa
formula1 = 'n1 ~ male'
results1 = smf.ols(formula1, data=df[df['REGION'] == 1]).fit()

print(results1.summary())
```

```python
# More complex model
formula2 = 'n1 ~ male + v012 + v136 + C(elect) + C(radio) + C(tv) + C(car) + C(educ) + C(marstat) + C(educp) + C(year) + C(iyr) + C(abc)'
results2 = smf.ols(formula2, data=df[(df['sextwin'] != 9999) & (df['REGION'] == 1)]).fit()

print(results2.summary())
```

# Central Africa Replication

```python
from patsy import dmatrices

# Define regions
central_codes = {'TD3', 'TD4', 'CM2', 'CM3', 'CM4', 'CF3', 'GA3'}
df['REGION'] = df['v000'].apply(lambda x: 2 if x in central_codes else 0)

# Filter data for Central Africa where sextwin is known
df_central = df[(df['REGION'] == 2) & (df['sextwin'].notna())]

# Simple regression analysis
results1 = smf.ols('n1 ~ male', data=df_central).fit()
print(results1.summary())
```

```python
# More complex model
results2 = smf.ols('n1 ~ male + v012 + v136 + C(elect) + C(radio) + C(tv) + C(car) + C(educ) + C(marstat) + C(educp) + C(year) + C(iyr) + C(abc)', data=df_central).fit()
print(results2.summary())
```

```python
# Fixed effects regression with clustering
y, X = dmatrices('n1 ~ male + v012 + v136 + C(elect) + C(radio) + C(tv) + C(car) + C(educ) + C(marstat) + C(educp) + C(year) + C(iyr) + C(abc) - 1', data=df_central, return_type='dataframe')
model_fe = sm.OLS(y, X)  # No constant added (-1 in formula)
fe_res = model_fe.fit(cov_type='cluster', cov_kwds={'groups': df_central['idtwin1b']})
print(fe_res.summary())
```

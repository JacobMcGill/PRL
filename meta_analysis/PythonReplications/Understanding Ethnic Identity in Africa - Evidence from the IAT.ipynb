```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
data_paths = [
    "Lowes_Nunn_Robinson_Weigel_AERPP_2015.dta"
]
```

```python
df = [pd.read_stata(path) for path in data_paths]
df = pd.concat(df)
# df.set_index('ID', inplace=True)
df
```

```python
# Delete observations with missing data 
# keep if missing(IAT)!=1 & missing(closeness)!=1 & missing(perception)!=1
df = df[ (df['IAT'].notnull()) & (df['closeness'].notnull()) & (df['perception'].notnull()) ]
```

```python
# closeness variable contains string variables 

# convert 'Closest together' to 5.0
df['closeness'] = df['closeness'].replace('Closest together', 5)

# convert 'Farthest apart' to 0.0
df['closeness'] = df['closeness'].replace('Farthest apart', 0)

# Convert 'closeness' variable from category type to int type
df['closeness'] = df['closeness'].cat.codes.astype(int)
```

```python
# Define and fit the fixed-effects regression model for feeling of closeness (col 1)

model = smf.ols("closeness ~ same_ethnicity + C(target_group) + C(ID)", df)
result = model.fit()

# Print the summary of the regression results
print(result.summary())

# 0.942 matches
```

```python
# Define and fit the fixed-effects regression model for positivity of perception (col 2)

model = smf.ols("perception ~ same_ethnicity + C(target_group) + C(ID)", df)
result = model.fit()

# Print the summary of the regression results
print(result.summary())

# 0.597 matches 
```

```python
# Define and fit the fixed-effects regression model for ST ethnicity IAT D-score (col 3)

model = smf.ols("IAT ~ same_ethnicity + C(target_group) + C(ID)", df)
result = model.fit()

# Print the summary of the regression results
print(result.summary())

# 0.062 matches
```

```python
# Define and fit the fixed-effects regression model for feeling of closeness (col 4)

model = smf.ols("closeness ~ same_ethnicity + enum_same_ethnicity + C(target_group) + C(ID)", df)
result = model.fit()

# Print the summary of the regression results
print(result.summary())

# 0.951 matches
```

```python
Trump=open('TrumpAIEO.txt','r') #file 1
Biden=open('BidenAIEO.txt','r') #file 2

file_list = [Trump, Biden] # This would hold all your open files
num_files = len(file_list)

frequencies = {} # We'll just make one dictionary to hold the frequencies

for i, f in enumerate(file_list): # Loop over the files, keeping an index i
    for line in f: # Get the lines of that file
        for word in line.split(): # Get the words of that file
            if not word in frequencies:
                frequencies[word] = [0 for _ in range(num_files)] # make a list of 0's for any word you haven't seen yet -- one 0 for each file

            frequencies[word][i] += 1 # Increment the frequency count for that word and file

print(frequencies)
```

```python
import nltk
from nltk import word_tokenize
from nltk.probability import FreqDist
import urllib.request
from matplotlib import pyplot as plt
from wordcloud import WordCloud
nltk.download('punkt')
```

```python
Trump=open('TrumpAIEO.txt','r').read()
Biden=open('BidenAIEO.txt','r').read()

Twords = word_tokenize(Trump)
Bwords = word_tokenize(Biden)
```

Trump AI EO

```python
Tfdist=FreqDist(Twords)
Tfdist.most_common(10)

#create an empty list to store words
Twords_no_punc = []

#iterate through the words list to remove punctuations
for word in Twords:
    if word.isalpha():
        Twords_no_punc.append(word.lower())

#find the frequency of words
Tfdist = FreqDist(Twords_no_punc)

#Plot the 10 most common words
Tfdist.plot(10)
plt.show()

#Download and import list of stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords

#list of stopwords
stopwords_list = stopwords.words("english")
print(stopwords_list)

#create an empty list to store clean words
Tclean_words = []

#Iterate through the words_no_punc list and add non stopwords to the new clean_words list
for word in Twords_no_punc:
    if word not in stopwords_list:
        Tclean_words.append(word)

print(f"The total number of words without punctuation and stopwords is {len(Tclean_words)}")
```

```python
#find the frequency of words
Tfdist = FreqDist(Tclean_words)

#Plot the 10 most common words
Tfdist.most_common(50)
```
